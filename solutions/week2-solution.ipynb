```python
#import necessary modules
import numpy as np 
from matplotlib import pyplot as plt 

# OK ... first WEEK 2 ... no loops/ifs as yet 
print('WEEK 2 \n') 

#Variable declarations 
#  q_in_swing    ..... fraction ... in-flow swings 
#  q_in_period   ..... [h]    ..... in-flow swings time period 
#  q_in, q_inAv  ..... [m3/h] ..... in-flow and average in-flow 
#  q_out         ..... [m3/h] ..... out-flow 
#  t             ..... [h]    ..... hours 
#  h, h_o        ..... [m]    ..... height and initial height 
#  h_max         ..... [m]    ..... tank maximum height 
#  A_tank        ..... [m^2]  ..... tank cross-sectional area 
#  V_tank        ..... [m^3]  ..... tank volume 
#  rho           ..... [kg/m^3] ... fluid density 
#  max_time      ..... [h]    ..... total modelling time 
#  h_set         ..... [m]    ..... control valve set point to avoid overflow 

#Data 
q_in_swing  =  0.4 
q_in_period =  6.0 

max_time = 24.0 
h_set  = 12 

h_o    = 10 
A_tank =  5 
V_tank = 75  
rho    = 1000 

q_inAv = 3500 
q_out  = q_inAv 

#Calculations 

# TASK 1 
print('   Task 1')

#first, the tank maximum height 
h_max = V_tank / A_tank 
print('     Max height of tank = ',h_max)

time = [5,10,15,20]
Qin5 = q_inAv*( 1 + q_in_swing*np.sin(np.pi*5/q_in_period))
Qin10 = q_inAv*( 1 + q_in_swing*np.sin(np.pi*10/q_in_period))
Qin15 = q_inAv*( 1 + q_in_swing*np.sin(np.pi*15/q_in_period))
Qin20 = q_inAv*( 1 + q_in_swing*np.sin(np.pi*20/q_in_period))

print('     Time [h] = ',time)
print('     Qin [kg/h] = ',Qin5,Qin10,Qin15,Qin20)
print('     This numbers do not show the true nature of the perturbation. This alone is not enough to study the problem \n')

#TASK 2 
print('   Task 2 ') 

# We define the properties of the concrete and g

Fy15 = 15000000
Fy20 = 20000000
Fy25 = 25000000

g = 9.81

# We assume that the thickness is constant, and should be enough to withstand the maximum possible amount of water (h_max)

# Thickness
t15 = 1.2*(rho*g*h_max**3)/Fy15
t20 = 1.2*(rho*g*h_max**3)/Fy20
t25 = 1.2*(rho*g*h_max**3)/Fy25

# Total area of concrete needed (assuming constant thickness) - For C15 (repeat for other concretes)
r_in = np.sqrt(A_tank/np.pi) #inner radius
r_out15 = r_in + t15# outer radius
Aout = np.pi*(r_out15**2)

# Total volume 
V15 = (Aout-A_tank)*h_max

# Total cost  
cost_C15 = V15*58

print('     Results for concrete C15 ') 
print('          Thickness C15 [m] = ',t15)
print('          Total volume needed of C15 [m3] = ',V15)
print('          Cost of C15 concrete [£] = ',cost_C15)

###############
# Concrete C20
##############
r_out20 = r_in + t20# outer radius
Aout20 = np.pi*(r_out20**2)

# Total volume 
V20 = (Aout20-A_tank)*h_max

# Total cost  
cost_C20 = V20*72

print('     Results for concrete C20 ') 
print('          Thickness C20 [m] = ',t20)
print('          Total volume needed of C20 [m3] = ',V20)
print('          Cost of C20 concrete [£] = ',cost_C20)

###############
# Concrete C25
##############
r_out25 = r_in + t25# outer radius
Aout25 = np.pi*(r_out25**2)

# Total volume 
V25 = (Aout25-A_tank)*h_max

# Total cost  
cost_C25 = V25*80

print('     Results for concrete C25 ') 
print('          Thickness C25 [m] = ',t25)
print('          Total volume needed of C25 [m3] = ',V25)
print('          Cost of C25 concrete [£] = ',cost_C25)
  
```
