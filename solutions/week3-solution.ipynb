```python
#import necessary modules
import numpy as np 
from matplotlib import pyplot as plt 

# OK ... now we have ifs/fors .... no functions nor plots yet 
print('WEEK 3 \n') 

#Variable declarations 
#  q_in_swing    ..... fraction ... in-flow swings 
#  q_in_period   ..... [h]    ..... in-flow swings time period 
#  q_in, q_inAv  ..... [m3/h] ..... in-flow and average in-flow 
#  q_out         ..... [m3/h] ..... out-flow 
#  t             ..... [h]    ..... hours 
#  h, h_o        ..... [m]    ..... height and initial height 
#  h_max         ..... [m]    ..... tank maximum height 
#  A_tank        ..... [m^2]  ..... tank cross-sectional area 
#  V_tank        ..... [m^3]  ..... tank volume 
#  rho           ..... [kg/m^3] ... fluid density 
#  max_time      ..... [h]    ..... total modelling time 
#  h_set         ..... [m]    ..... control valve set point to avoid overflow 

#Data 
q_in_swing  =  0.4 
q_in_period =  6.0 

max_time = 24.0 
h_set  = 12 

h_o    = 10 
A_tank =  5 
V_tank = 75  
rho    = 1000 

q_inAv = 3500 
q_out  = q_inAv 

#Calculations 

# Task 1 
print('   Task 2: control valve working')

#first, the tank maximum height 
h_max = V_tank / A_tank 
print('     Max height of tank = ',h_max)

#preparations 
time = [] 
q_in = []
h    = []
level_max = None 
out_open  = False 
overflow  = False
t_switch    = 0.0 
print('     t [h]      Q_in [Kg/h]        h [m]')

#loop for calculating height changes over 24 hours 
for n in range(0,int(max_time)*4,1):
    #the time 
    time.append(n*0.25)
    #the in-flow and height 
    inflow    = q_inAv*( 1 + q_in_swing*np.sin(np.pi*time[n]/q_in_period)) 
    q_in.append(inflow)
    if n == 0:
        h_old = h_o 
        level_max = h_o 
        q_out = 0.0 
        h.append(h_o)
    else:
        #the tank level 
        level = h_old + (q_inAv-q_out)/(rho*A_tank)*(time[n]-t_switch) \
        + (q_in_swing*q_inAv/(rho*A_tank))*(q_in_period/np.pi)*(1-np.cos(np.pi*time[n]/q_in_period))
        h.append(level) 
        #open/close the valve depending on whether the level is above/below set-point 
        if level > h_set and out_open == False:
            h_old = h[n]
            t_switch = time[n]
            q_out = q_inAv 
            out_open = True 
        elif level < h_set and out_open == True:
            h_old = h[n]
            q_out = 0.0 
            t_switch = 0.0 
            out_open = False
    #check the highest level in the tank 
    if h[n] > level_max: 
        level_max = h[n]
    #warn imminent overflow 
    if h[n] >= (h_max-1):
        print('     ATT: overflow is imminent! present level [m] = ',h[n]) 
        overflow = True 
        break 
    #show results every hour
    if n % 4 == 0:
        print('   ',"{:.2f}".format(time[n]),'  ',"{:.2f}".format(q_in[n]),'  ',"{:.2f}".format(h[n])) 

#Report the maximum fluid level in the tank 
if overflow:
    print(' .... ')
else: 
    print('\n Maximum level in the tank within safety limits, max_level [m] = ',"{:.2f}".format(level_max))
    
```

```python
# Recalculating the cost of concrete per metre high of the wall

Fy = 15000000 # Choose one of the following options: 15000000, 20000000, or 25000000

g = 9.81
r_in = np.sqrt(A_tank/np.pi) #inner radius

thickness = [] # Creating empty lists to fill in for loop
h = []
Vol = []

if Fy == 15000000 or Fy == 20000000 or Fy == 25000000: # We only want to do calculations for these 3, if not, don't do anything!
    
    # We assume that the thickness varies each metre

    for i in range(1,int(h_max)+1):
        t = 1.2*(rho*g*i**3)/Fy # Thickness
        r_out = r_in + t # outer radius
        Aout = np.pi*(r_out**2) # Area of concrete
        V = (Aout-A_tank)*1 # Volume at every step (in every metre depth of wall)
        
        thickness.append(t) # Filling up lists in every step
        h.append(i)
        Vol.append(V)
        
    Volume = sum(Vol)
        
    cost = []
    if Fy==15000000:
        cost = Volume*58 # Total cost 
    elif Fy == 20000000:
        cost = Volume*72
    elif Fy == 25000000:
        cost = Volume*80


    print("Depth of water (m)     Minimum wall thickness (m)     Volume in that section (m3)")

    for i in range(0,int(h_max)): 
        print("     ",h[i],"                       ","{:.3f}".format(thickness[i]),"                         ", "{:.3f}".format(Vol[i]))
        
    print ("The total volume of concrete needed is", "{:.3f}".format(Volume), "m3.")
    print("The cost of concrete is £","{:.2f}".format(cost))

else:
    print("There is no cost data for the type of concrete selected. Please choose one of the following Fy: 15000000, 20000000 or 25000000 Pa.")

    
# There are several ways to do the next part. One is by "integrating" a bunch of small
# cylinders as each section of wall tapers.

extra_volume = []
for i in range(0,int(h_max)): # 
    # Consider each 1 meter
    if i > 0:
        start_thickness = thickness[i-1]
    else: 
        start_thickness = 0

    end_thickness = thickness[i]
    
    # Create a bunch of small cyclinders of thickness ranging linearly between these
    # 1000 should be enough for a good estimate
    
    num_points = 3
    T = np.linspace(start_thickness, end_thickness, num_points)
    
    dh = 1/num_points # Each cylinder has a height of 1 meter/num_points
    
    vol = 0 # Keep a total of the additional volume
    for t in T:
        vol += (np.pi*t**2 - np.pi*start_thickness**2) * dh # New vol minus existing vol
        
    extra_volume.append(vol)
    
    

Final_volume_linear = Volume + sum(extra_volume)
print("To build a wall with a linear profile, we need", "{:.3f}".format(Final_volume_linear),"m3 of concrete.")
    
```
